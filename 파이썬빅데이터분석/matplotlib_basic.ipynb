```python
years = df.index.values
years
counts = df['trans_count'].values  #컬럼에 있는 데이터들
counts
# plt.plot(years, counts) # 올바른 정보를 알수 없다
# plt.show()   # 1970~200년 사이의 변화량이 매우 작음 

plt.plot(years, np.log(counts)) # 로그 사용
plt.show() 
```

```python
e = 2.71828182845904   #오일러 상수, 자연로그의 밑
np.log(e)  #자연로그 사용

a = np.log(1000)
print(a) #if 상용로그는 3이엇을 것인데 자연로그로 햇기때문에 6.9가나옴
np.exp(a) # 주로 로그취한걸 제곱해서 다시 되돌려줌 
```

```python
poly = np.polyfit(years,np.log(counts), deg = 1) #학습 #deg는 몇차방정식인지 알수잇음
#deg= degree :차수
# polyfit(x,y,deg) : x값과 y값을 사용하여 기울기를 구함
#Least squares polynomial fit
#선과 데이터가 가장 적은 선   다항식의 적합도(polynomial fit)
print(type(poly))
print('poly',poly[0])   #\, RLDNFRL 
print('poly',poly[1])   #b, y절편
print('Poly',poly)
plt.semilogy(years,counts,'o')  #semilogy: log값으로 나오게 한다 
plt.semilogy(years,np.exp(np.polyval(poly,years)),'o')   #189page #exp: 로그값을 지수형태로 바꾼는것(로그의 반대값 ) 
plt.show()  #poly:다항식
# Hypothesis 가설, 예측 방정식 
# f(x) =  0.3615592104293512 *x -705.7831945013617
# predict = np.exp((fx))
plt.semilogy(years,np.exp(np.polyval(poly,years)))  # w
predict = np.exp(np.polyval(poly,[2015,2016,2017,2019,2020]))  #예측
print('예측')
print('2015:',predict[0])
print('2019:',predict[3])
print('2020:',predict[4])
```

범례의 주석

```python
df = pd.read_csv('transcount.csv')
gpu = pd.read_csv('gpu_transcount.csv')

df = df.groupby('year').agg(np.mean)
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu, how='outer', left_index=True,
             right_index=True)

df = df.replace(np.nan,0) # 결측치를 0으로 수정
years = df.index.values
counts = df['trans_count'].values 
gpu_counts = df['gpu_trans_count'].values

cnt_log = np.log(counts)

gpu_start = gpu.index.values.min()  #1997
y_ann = np.log(df.at[gpu_start,'trans_count'])   #at이머양?
ann_str = 'First GPUWn %d'% gpu_start
print(ann_str)

plt.annotate(ann_str,xy=(gpu_start,y_ann),
             arrowprops=dict(arrowstyle='->'),
             xytext=(-30,+70),
             textcoords='offset points')

poly = np.polyfit(years,cnt_log,deg=1)
plt.plot(years,np.polyval(poly,years),label='Poly Fit')

# c: color, s: size, apha: 투명도
plt.scatter(years,cnt_log, c = 200*years,
            s= 20 + 200*gpu_counts/gpu_counts.max(),
            alpha = 0.5,label = 'Scatter Plot')  # 버블차트
#plt.plot(years,cnt_log)

plt.legend(loc='lower right')
plt.grid(True)
plt.xlabel('year')
plt.ylabel('trans_count')
plt.title('trans/gpu counts per year')
plt.show()
```

```python
#직선에 주석 달기 
x = np.arange(20)
y = x *2
y2 = (x-10) **2
x_start = x.mean()
y_start = x_start *2
ann_str =  'X mean \n %d' %x_start
plt.plot(x,y, label = '직선')
plt.plot(x,y2, label  = '포물선')
plt.legend(loc = 'upper center')
plt.annotate(ann_str, xy= (x_start, y_start),
             arrowprops = dict(arrowstyle = '->'),
            xytext = (-30,+70), textcoords = 'offset points')
plt.title('직선에 주석/범례달기')
x_start = 10
y_start = 0
ann_str = '기울기\n %d' %y_start
plt.annotate(ann_str, xy= (x_start, y_start),
             arrowprops = dict(arrowstyle = '->'),
            xytext = (+30,+70), textcoords = 'offset points')

plt.show()
```

```python
# 팬더스의 plotting 사용~
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)
gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu, how = 'outer', left_index = True, right_index = True)

df['trans_count'] = np.log(df['trans_count'].values)
df['gpu_trans_count'] = np.log(df['gpu_trans_count'].values)

df = df.replace(np.nan,0)   # 결측치 0

df.plot.box()
plt.show()
```

```python
# matplotlib 사용
plt.boxplot([df['trans_count'], df['gpu_trans_count']],    # []왜 이걸로 묶어야하죠?
            labels = ['trans_count','gpu_trans_count'])
plt.text(1,df['trans_count'].median(), round(df['trans_count'].median()))          # 첫번째:위치, 두번째:값
plt.text(2,df['gpu_trans_count'].median(), df['gpu_trans_count'].median())         # 첫번째:위치, 두번째:값
```
