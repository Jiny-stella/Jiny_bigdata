4.
- Boston Housing Data Set 열의 통계 출력
  boston_train.csv 파일을 각 열(총 10개)에 대한 통계를 출력하는 코드를 구현하세요
  sum,mean,median,min,max를 출력시킨다


```python
# 4
df = pd.read_csv('boston_train.csv')  #파일을 불러오면 데이터 프레임 방식으로 데려오는것인가?
df_col = df.iloc[:,:10]
print(df_col)
print(df_col.sum())
print(df_col.mean())
print(df_col.median())
print(df_col.min())
print(df_col.max())

for col in df.columns:
    print('[',col,']')
    print('sum:',df[col].sum())
    print('mean:',df[col].mean())
    print('median:',df[col].median())
    print('min:',df[col].min())
    print('max:',df[col].max())
    print('describe:',df[col].describe())
```

 아래 두개의 데이터 프레임을 만들어 merge()로 통합하세요(inner)

```python
product_ab = {'판매월':['1월','2월','3월','4월'],'제품A':[100,150,200,130],'제품B':[90,110,140,170]}
ab = pd.DataFrame(product_ab)
product_cd = {'판매월':['1월','2월','3월','4월'],'제품C':[112,141,203,134],'제품D':[90,110,140,170]}
cd = pd.DataFrame(product_cd)

mer = pd.merge(ab,cd, how = 'inner',on = '판매월')
mer
```

```python
key_l = {'key':['A','B','C'],'left':[1,2,3]}
key_r = {'key':['A','B','D'],'right':[4,5,6]}
df_l = pd.DataFrame(key_l)
df_r = pd.DataFrame(key_r)
# 8-1
m1 = pd.merge(df_l,df_r, how ='inner',on = 'key')
m1
m2 = pd.merge(df_l,df_r, how ='outer',on = 'key')
m2
m3 = pd.merge(df_l,df_r, how ='left',on = 'key')
m3
m4 = pd.merge(df_l,df_r, how ='right',on = 'key')
m4
```
